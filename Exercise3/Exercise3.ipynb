```python
df = df[['Survived', 'Age', 'Sex', 'Pclass']]
df = pd.get_dummies(df, columns=['Sex', 'Pclass'])
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('Survived', axis=1)
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

## Generate Results

```python
# Reload the sample submission file to ensure 'PassengerId' is available
sample_submission_url ='https://github.com/robitussin/CCMACLRL_EXERCISES_OLD/blob/f5fdfba1aabbbafcf2fe49ca6e1c8f92f69d1458/datasets/titanic/gender_submission.csv?raw=true'
sf = pd.read_csv(sample_submission_url)
id = sf.pop('PassengerId')

# Preprocess the test data (dt) similar to the training data (df)
dt = dt[['Age', 'Sex', 'Pclass', 'Fare']]

# Handle missing values in Age and Fare
dt['Age'].fillna(dt['Age'].median(), inplace=True)
dt['Fare'].fillna(dt['Fare'].median(), inplace=True)

# Convert categorical features to numerical
dt = pd.get_dummies(dt, columns=['Sex', 'Pclass'])

train_cols = x_train.columns
test_cols = dt.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    dt[c] = 0

dt = dt[train_cols]

# Make predictions on the preprocessed test data
y_pred = model.predict(dt)

# Create a submission DataFrame
submission_df = pd.DataFrame({
    'PassengerId': id,
    'Survived': y_pred
})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission_file.csv', index=False)
print("Submission file created: submission_file.csv")
```
